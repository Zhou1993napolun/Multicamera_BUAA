```python
qpath = query_paths[0]
q_pids = int(qpath[:4])
q_camids = []
for c in qpath[10:]:
    q_camids.append(int(c))
```

```python
# from collections import defaultdict
g_indices = {}
g_paths = {}
g_pids = {}
cnt = 0
cnt_dict = {}
for camid in q_camids:
    cnt_dict[camid] = 0
for g_idx, gpath in enumerate(gallery_paths):
    camid = int(gpath[6])
    if camid not in cnt_dict:
        continue
    time = gpath[9:13]
    if time not in g_indices:
        g_indices[time] = {}
    if camid not in g_indices[time]:
        g_indices[time][camid] = [cnt_dict[camid]]
    else:
        g_indices[time][camid].append(cnt_dict[camid])
    if time not in g_paths:
        g_paths[time] = {}
    if camid not in g_paths[time]:
        g_paths[time][camid] = [gpath]
    else:
        g_paths[time][camid].append(gpath)
    if time not in g_pids:
        g_pids[time] = {}
    if camid not in g_pids[time]:
        g_pids[time][camid] = [int(gpath[:4])]
    else:
        g_pids[time][camid].append(int(gpath[:4]))
    cnt_dict[camid] += 1
```

```python
g_dists = {}
for time, cam_dict in g_indices.items():
    if time not in g_dists:
        g_dists[time] = {}
    for camid, indice in cam_dict.items():
        index = q_camids.index(camid)
        tmp_array = datas[index][:, indice]
        g_dists[time][camid] = tmp_array
```

```python
def judge_shape(g_dists, g_paths):
    try:
        for time, dist_cam_dict in g_dists.items():
            path_cam_dict = g_paths[time]
            for dist_cam, arr in dist_cam_dict.items():
                if not arr.shape[1] == len(path_cam_dict[dist_cam]):
                    raise
    except:
        print('sth wrong')
judge_shape(g_dists, g_paths)
```

```python
def sort_rows_and_get_indices(array):
    # 获取每一行排序后的索引
    # sorted_indices = np.argsort(array, axis=1)[:, ::-1]
    sorted_indices = np.argsort(array, axis=1)


    # 对每一行进行排序
    sorted_rows = np.take_along_axis(array, sorted_indices, axis=1)

    return sorted_rows, sorted_indices
```

```python
g_results = {}
for time, cam_dict in g_dists.items():
    if time not in g_results:
        g_results[time] = {}
    for cam, dist in cam_dict.items():
        sorted_dist, sorted_indice = sort_rows_and_get_indices(dist)
        rank0_pos = sorted_indice[0][0]
        preci = (int(pid) == g_pids[time][cam][rank0_pos])
        r0path = g_paths[time][cam][rank0_pos]
        r0dist = sorted_dist[0][0]
        g_results[time][cam] = [r0path, str(preci), str(r0dist)]
```
