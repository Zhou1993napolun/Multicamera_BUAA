```python
import json
import numpy as np
import os
import shutil
from collections import defaultdict
```

```python
root = '../orange_demo/'
```

```python
def baseline_match(root, src='baseline', dst='matches', k=100):
    root_folder = os.path.join(root, src)
    match_dict = os.path.join(root_folder, 'matches.json')
    storage_folder = os.path.join(root_folder, 'matches')
    log_file = os.path.join(storage_folder, 'matches_logs.json')
    log_dict = defaultdict(str)
    with open(match_dict, 'r') as f:
        datas = json.load(f)
    cnt = 0
    for q, gs in datas.items():
        if cnt >= k:
            break
        r0 = gs[0]
        if not r0[1]:
            cnt += 1
            query_name = q.split('/')[-1]
            gallery_name = r0[0].split('/')[-1]
            path = os.path.join(storage_folder, query_name[:-4])
            if not os.path.exists(path):
                os.makedirs(path)
                print(f'The directory {path} has been created')
            else:
                print(f'The directory {path} has already existed, the following operation will repeat')
            query_src = os.path.join('..', q)
            query_dst = os.path.join(path, query_name)
            shutil.copy2(query_src, query_dst)
            gallery_src = os.path.join('..', r0[0])
            gallery_dst = os.path.join(path, gallery_name)
            shutil.copy2(gallery_src, gallery_dst)
            log_dict[query_name] = gallery_name
    with open(log_file, 'w') as json_flie:
        json.dump(log_dict ,json_flie, indent=4)
            

    # return datas
baseline_match(root)
```
