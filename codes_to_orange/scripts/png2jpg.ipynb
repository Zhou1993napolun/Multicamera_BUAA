```python
from PIL import Image
import os
from os.path import join
from tqdm import tqdm

def png_to_jpg(input_path, output_path, quality=95):
    """
    将PNG图像转换为JPG格式
    :param input_path: 输入PNG文件路径
    :param output_path: 输出JPG文件路径
    :param quality: JPG图像质量（0-100，默认为95）
    """
    try:
        # 打开PNG图像
        imgs = os.listdir(input_path)
        # print(imgs)
        for img in imgs:
            img_input = join(input_path, img)
            img_output = join(input_path, img.split('.')[0]+'.jpg')
            # print(img_input)
            # print(img_output)
            if os.path.exists(img_output):
                continue
            with Image.open(img_input) as i:
                # 转换并保存为JPG格式
                i.convert('RGB').save(img_output, 'JPEG', quality=quality)
        # print(f"转换成功: {output_path}")
    except Exception as e:
        print(f"转换失败: {e}")
```

```python
root = '../datasets/MultiviewX_eval/gallery_mmoe'
files = os.listdir(root)
for file in tqdm(files):
    file_path = join(root, file)
    png_to_jpg(file_path, file_path)
```
