```python
import json
import os
from os.path import join
```

```python
root = '../datasets/'
ori_dataset = 'Wildtrack_eval'
reranked_dataset = 'Wildtrack_eval_rerank'
ori_path = join(root, ori_dataset)
reranked_path = join(root, reranked_dataset)
```

```python
ori_num_set = set()
reranked_num_set = set()
```

```python
def fulfill_set(s, root, qp = 'query', gp = 'gallery_mmoe'):
    qnames = os.listdir(join(root, qp))
    for qname in qnames:
        qid = int(qname[:4])
        s.add(qid)
    gnames = os.listdir(join(root, gp))
    for gname in gnames:
        gid = int(gname[:4])
        s.add(gid)
fulfill_set(ori_num_set, ori_path)
fulfill_set(reranked_num_set, reranked_path)
```
