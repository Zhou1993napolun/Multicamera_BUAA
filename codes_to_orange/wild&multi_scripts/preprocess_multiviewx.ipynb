```python
# train, test, val 对应 train, gallery, query
def file_transfered(source, target, ref):
    img_groups_all = sorted(os.listdir(source))
    ref_train = sorted(os.listdir(join(ref, 'train')))
    ref_gallery = sorted(os.listdir(join(ref, 'test')))
    ref_query = sorted(os.listdir(join(ref, 'val')))
    candidats = [ref_train, ref_gallery, ref_query]
    targets = ['train', 'test', 'val']
    img_cnt = 0
    file_cnt = 0
    for tar, img_files in tqdm(zip(targets, candidats)):
        target_file = join(target, tar)
        full = join(target_file, 'full')
        if not os.path.exists(full):
            os.makedirs(full)
        for img_file in img_files:
            source_file = join(source, img_file)
            imgs = os.listdir(source_file)
            for img_name in imgs:
                camid = img_name[6]
                sub_file = join(target_file, camid)
                if not os.path.exists(sub_file):
                    os.makedirs(sub_file)
                source_full_path = join(source_file, img_name)
                target_full_path = join(sub_file, img_name)
                img_cnt += 1
                shutil.copy2(source_full_path, target_full_path)
            shutil.copytree(source_file, join(full, img_file))
            file_cnt += 1
        print(f'{file_cnt} files has been processed')
        print(f'{img_cnt} files has been processed')
            
    # for dir in dirs:
    #     dir_path = os.path.join(root, dir)
    #     imgs = os.listdir(dir_path)
    #     for img in imgs:
    return img_groups_all, ref_train, ref_gallery, ref_query 
```

```python
def judge_is_injective(query_path,gallery_path):
    q_id_set = set()
    g_id_set = set()
    query_files = os.listdir(query_path)
    gallery_files = os.listdir(gallery_path)
    # if os.path.isfile(query_files[0]):
    for query_name in query_files:
        q_id = int(query_name[:4])
        q_id_set.add(q_id)
    for gallery_name in gallery_files:
        g_id = int(gallery_name[:4])
        g_id_set.add(g_id)
    missed = set()
    num_missed = 0
    for q in q_id_set:
        if q not in g_id_set:
            missed.add(q)
            num_missed += 1
    query_dirname = os.path.basename(query_path)
    print(f'for {query_dirname}, {num_missed} ids can not be found')
    return missed
    # elif os.path.isdir(query_files[0]):
    #     for query_file in query_files:
    #         q_id = int(query_file[:4])
```

```python
query_root = join(root, 'eval_format', 'val')
gallery_root = join(root, 'eval_format', 'test')
query_paths = [join(query_root, subpath) for subpath in os.listdir(query_root)]
gallery_paths = [join(gallery_root, subpath) for subpath in os.listdir(query_root)]
```

```python
misses = []
query_paths.sort()
gallery_paths.sort()
for qp, gp in zip(query_paths, gallery_paths):
    miss = judge_is_injective(qp, gp)
    misses.append(miss)
```

```python
def mv_files(src, dst):
    files = os.listdir(src)
    for file in files:
        file_path = join(src, file)
        imgs = os.listdir(file_path)
        for img in imgs:
            img_path = join(file_path, img)
            shutil.copy2(img_path, join(dst, img))
```

```python
cur_root = '../datasets/MultiviewX_eval/'
cur_dst = '../datasets/MultiviewX2/'
srcs = [join(cur_root, 'train', 'full'), join(cur_root, 'query', 'full'), join(cur_root, 'gallery_mmoe', 'full')]
dsts = [join(cur_dst, 'bounding_box_train'), join(cur_dst, 'query'), join(cur_dst, 'bounding_box_test')]
for src, dst in zip(srcs, dsts):
    mv_files(src, dst)
```
